```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv(r"taxi_trip_pricing.csv")
data.head()
```

```python
data.isnull().sum()
```

```python
data=data.dropna()
```

```python
data.isnull().sum()
```

```python
data.duplicated().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
data.columns
```

```python
categorical_data=['Time_of_Day','Day_of_Week','Traffic_Conditions','Weather']
```

```python
for i in categorical_data:
    print(i)
    print(data[i].unique())
    print("\n")
```

```python
for i in categorical_data:
    data.loc[:,i]=le.fit_transform(data.loc[:,i])
```

```python
data.head()
```

```python
x=data.drop(['Trip_Price'],axis=1)
y=data['Trip_Price']
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(data.corr(),annot=True)
plt.show()
```

```python
j=1
plt.figure(figsize=(20,10))
for i in x:
    plt.subplot(2,5,j)
    sns.boxplot(y=i,data=data)
    plt.xticks(rotation=90)
    j+=1
plt.show()
```

```python
# capping to outliers to remove them.
def cap_outliers(df,columns):
    q1=df[columns].quantile(0.25)
    q3=df[columns].quantile(0.75)
    IQR=q3-q1
    lb=q1-1.5*IQR
    ub=q3+1.5*IQR
    df[columns]=df[columns].apply(lambda a : lb if a<lb else ub if a>ub else a)
cap_outliers(data,'Trip_Distance_km')
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
# Scaling data
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
```

```python
x_train_scaled=ss.fit_transform(x_train)
x_train_scaled=pd.DataFrame(x_train_scaled,columns=x.columns)
```

```python
x_test_scaled=ss.transform(x_test)
x_test_scaled=pd.DataFrame(x_test_scaled,columns=x.columns)
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
r=[]
mae=[]
mse=[]
algo={'knn':KNeighborsRegressor(n_neighbors=6),
      'linear':LinearRegression(),
      'svr':SVR(),
      "rf":RandomForestRegressor() 
     }

for name,model in algo.items():
    model.fit(x_train_scaled,y_train)
    y_pred=model.predict(x_test_scaled)
    rs=r2_score(y_test,y_pred)
    maes=mean_absolute_error(y_test,y_pred)
    mses=mean_squared_error(y_test,y_pred)
    r.append(rs)
    mae.append(maes)
    mse.append(mses)
print("\nr2_score: ",r)
print("mean_absolute_error: ",mae)
print("meas_squared_error: ",mse)
```

```python
fig,axes=plt.subplots(1,3,figsize=(10,4))
axes[0].plot(algo.keys(), r, marker='D', color='blue')
axes[0].set_title("R Score")

# Plot on second subplot
axes[1].plot(algo.keys(), mae, marker='D', color='green')
axes[1].set_title("MAE")

# Plot on third subplot
axes[2].plot(algo.keys(), mse, marker='D', color='red')
axes[2].set_title("MSE")
plt.tight_layout()
plt.show()
```

```python
# svr is selected as per aboce graphs
```

```python
from sklearn.ensemble import RandomForestRegressor
r2_sc=[]
msee=[]
maee=[]
for j in range(4,50):
    rf=RandomForestRegressor(n_estimators=j)
    rf.fit(x_train_scaled,y_train)
    y_p=rf.predict(x_test_scaled)
    r2_sc.append(r2_score(y_test,y_p))
    maee.append(mean_absolute_error(y_test,y_p))
    msee.append(mean_squared_error(y_test,y_p))
```

```python
plt.figure(figsize=(20,10))
plt.title("R2_Score of random forest")
plt.plot(r2_sc,marker='D')
plt.xticks(range(1,46))
plt.grid()
```

```python
plt.figure(figsize=(20,10))
plt.title("MAE of random forest")
plt.plot(maee,marker='D')
plt.xticks(range(1,46))
plt.grid()
```

```python
plt.figure(figsize=(20,10))
plt.title("MSE of random forest")
plt.plot(msee,marker='D')
plt.xticks(range(1,46))
plt.grid()
```
